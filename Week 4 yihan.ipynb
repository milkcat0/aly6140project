```python
# load packages
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
# Will allow us to embed images in the notebook
%matplotlib inline
```

```python
# load data sheet 1
efw_index_2019_df = pd.read_excel(
    'efw-2019-master-index-data-for-researchers.xlsx', 
    sheet_name='EFW Index 2019 Report', #name of the sheet
    header=4,
    index_col=1
)

efw_index_2019_df = efw_index_2019_df.loc[:, ~efw_index_2019_df.columns.str.contains('^Unnamed')]

efw_index_2019_df
```

```python
# load data sheet 2

efw_panel_data_2019_df = pd.read_excel(
    'efw-2019-master-index-data-for-researchers.xlsx', 
    sheet_name='EFW Panel Data 2019 Report', #name of the sheet
    header=2,
    index_col=1
)
efw_panel_data_2019_df = efw_panel_data_2019_df.loc[:, ~efw_panel_data_2019_df.columns.str.contains('^Unnamed')]

efw_panel_data_2019_df
```

```python
# load data sheet 3

raw_df = pd.read_excel(
    'efw-2019-master-index-data-for-researchers.xlsx',
    sheet_name='EFW 1950-1965',  #name of the sheet
    header=1,
    index_col=1
)

efw_historical_df = raw_df[['Country', 'EFW', 'Rank']]

efw_historical_df
```

```python
def mov_median(df,
               country_name: str,
               column_name: str,
               n: int):
    if n <= 0:
        raise ValueError("N must be non-negative and non-zero integer value")
    # FIXME: refactor into a more dynamic way than this temp solution
    try:
        df = df.loc[df['Country'] == country_name]
    except:
        df = df.loc[df['Countries'] == country_name]

    n = df.shape[0] if df.shape[0] < n else n
    new_df = df[column_name].rolling(n).median()
    # return new_df.dropna().tolist()[0]
    return new_df.dropna().to_frame()
```

```python
# mitr - marginal_income_tax_rate
top_20_mitr_series = efw_index_2019_df.groupby(['Countries'])['Top marginal income tax rate'].max()
top_20_mitr_df = top_20_mitr_series.sort_values(ascending=False)[: 20].to_frame()

plot = top_20_mitr_df.plot.bar(title='Top 20 country marginal income tax rates',
                               rot=0,
                               figsize=(25,5))
plot.set_ylabel('Marginal income tax rate')

plot
```

'''

```python
# plot a time series graph
uk_efw_df = mov_median(df=efw_panel_data_2019_df,
                       country_name='United Kingdom',
                       column_name='EFW',
                       n=3)
# uk_efw_df
plot = uk_efw_df.plot.line(title='3-Period moving median "EFW" for United Kingdom')
plot.set_ylabel('EFW')
plot.set_xlabel('Year')
plot
```

We often see patterns or relationships in scatterplots. A scatter plot shows a positive trend if y tends to increase as x increases.And strength refers to the degree of "scatter" in the plot. If the dots are widely spread, the relationship between variables is weak. If the dots are concentrated around a line, the relationship is strong.

In the scatter plot we got, the points cluster in a band is running from lower left to upper right and their dots are around a line,it means there is a strong positive correlation between two variables.When "Reliability of Police' increases, the Business Cost of Crime increases too.

A time series graph lines displays values against time. For the second graph,we can see there is a sharp rise of Economic Freedom Index from 1970 to 1995 and then curve tends to rise steadily from 1995 to 2003. And a sharp decline happened.
